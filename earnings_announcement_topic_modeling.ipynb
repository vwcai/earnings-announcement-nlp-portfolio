# Earnings Announcement Topic Modeling
This portfolio-oriented notebook presents a selected workflow for cleaning earnings announcement text, extracting topic features, and studying how disclosure language relates to short-window stock return reactions.

## Load and Clean Earnings Announcement Text

```python
# Load the earnings announcement dataset
import pandas as pd

df = pd.read_csv("ExpTask2Data.csv.gz")
print(f"The dataset has {df.shape[0]} rows.")
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import re


TABLE_NUMBER_RE = re.compile(
    r'^\s*'                       # Leading whitespace at the start of the string
    r'(?:'                        # Start of non-capturing group, matches two cases
        r'\$?\(?\s*'              # Case 1: Dollar sign outside optional left parenthesis
        r'-?'                     # Optional negative sign
        r'(?:\d{1,3}(?:,\d{3})*|\d+)' # Number part: digits with optional comma separators, or plain digits
        r'(?:\.\d+)?'             # Optional decimal part
        r'\s*\)?'                 # Optional right parenthesis
    r'|'                          # OR
        r'\(?\$?\s*'              # Case 2: Optional left parenthesis, dollar sign can be inside
        r'-?'                     # Optional negative sign
        r'(?:\d{1,3}(?:,\d{3})*|\d+)' # Number part: digits with optional comma separators, or plain digits
        r'(?:\.\d+)?'             # Optional decimal part
        r'\s*\)?'                 # Optional right parenthesis
    r')'
    r'\s*$'                       # Trailing whitespace at the end of the string
)

def replace_table_numbers_precise(line):
    words = line.split()
    new_words = []

    for i, word in enumerate(words):
        if not TABLE_NUMBER_RE.match(word):
            new_words.append(word)
            continue

        prev_ok = (i > 0 and TABLE_NUMBER_RE.match(words[i-1]))
        next_ok = (i < len(words)-1 and TABLE_NUMBER_RE.match(words[i+1]))

        if prev_ok or next_ok:
            new_words.append(" financial_number ")
        else:
            new_words.append(word)

    return " ".join(new_words)

def normalize_table_numbers(line):
    # delete whitespaces in the parenthesis that only contains financial digits
    line = re.sub(
        r'\(\s*([+-]?\$?-?(?:\d{1,3}(?:,\d{3})*|\d+)(?:\.\d+)?)\s*\)',
        r'(\1)',
        line
    )
    # delete whitespace between $ and the next digits 
    line = re.sub(
        r'\$\s*(\(?-?(?:\d{1,3}(?:,\d{3})*|\d+)(?:\.\d+)?\)?)',
        r'$\1',
        line
    )
    return line


# 1. $ + digits + unit
RE_DOLLAR_MAGNITUDE = re.compile(
    r'\(?\s*\$-?\d{1,3}(?:,\d{3})*(?:\.\d+)?(?:\s*(?:million|billion|trillion))?\s*\)?',
    re.IGNORECASE
)

# 2. digits + unit
RE_MAGNITUDE = re.compile(
    r'\(?\s*-?\d{1,3}(?:,\d{3})*(?:\.\d+)?(?:\s*(?:million|billion|trillion))\s*\)?',
    re.IGNORECASE
)

# 3. digits + % 
RE_PERCENT = re.compile(
    r'\(?\s*-?\d{1,3}(?:,\d{3})*(?:\.\d+)?%\s*\)?'
)

# 4. digits + shares (words are allowed between digits and shares)
RE_SHARES = re.compile(
    r'\(?\s*\$-?\d{1,3}(?:,\d{3})*(?:\.\d+)?(?:\s+\w+)*\s+(?:share|shares)\s*\)?',
    re.IGNORECASE
)

def replace_remaining_financial_numbers(line):
    line = RE_SHARES.sub(" financial_number ", line)
    line = RE_DOLLAR_MAGNITUDE.sub(" financial_number ", line)
    line = RE_MAGNITUDE.sub(" financial_number ", line)
    line = RE_PERCENT.sub(" financial_number ", line)
    return line

def clean_financial_number_spacing(text):
    # combine space before and after "financial_number"
    text = re.sub(r'\s*financial_number\s*', ' financial_number ', text)
    # delete extra space in the beginning
    text = re.sub(r'^\s+', '', text)
    text = re.sub(r'\s+$', '', text)
    # delete extra space
    text = re.sub(r' {2,}', ' ', text)
    return text

def replace_numbers_in_text(line_text):
    new_lines = []

    for line in line_text.splitlines():
        # just in case there are multiple lines
        line = normalize_table_numbers(line)
        if re.search(r'\d', line):
            line = replace_table_numbers_precise(line)
            line = replace_remaining_financial_numbers(line)
            line = replace_table_numbers_precise(line)
        line = clean_financial_number_spacing(line)
        new_lines.append(line)
    return "\n".join(new_lines)



def clean_sentences(txt):
    clean_txt = replace_numbers_in_text(txt)
    good_sents = []
    for sent in sent_tokenize(clean_txt):
        good_tokens = []
        nFinancialNumber = 0
        for word in word_tokenize(sent):                
            # filter out non-alpha or non-financial_number, building list of "good tokens"
            if not (word.isalpha() or word == 'financial_number'):
                continue
            if word == 'financial_number':
                nFinancialNumber += 1
            
            good_tokens.append(word)
        nGoodTokens = len(good_tokens)
        # Now add logic to check if there are sufficient tokens to keep the sentence:
        if nGoodTokens >= 5 and (nFinancialNumber / nGoodTokens) <= 0.5:
            good_sents.append(" ".join(good_tokens))
       
    return " ".join(good_sents) 
```

```python
df['clean_sent'] = df['ea_text'].apply(clean_sentences)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords

stops = stopwords.words('english')

# Update with parameters needed per instructions; I've provided a token_pattern
vec = CountVectorizer(token_pattern = r'\b[a-zA-Z_]{3,}[a-zA-Z]*\b',
                      ngram_range=(1,2),
                      max_features=1000,
                      stop_words=stops,
                      max_df=0.5) 

dtm = vec.fit_transform(df['clean_sent'])
```

## Fit and Evaluate Topic Models

```python
from sklearn.decomposition import LatentDirichletAllocation as LDA
from tmtoolkit.topicmod.evaluate import metric_coherence_gensim
import numpy as np

# Select a sample of 5000 for training; use random_state=123
sample = pd.DataFrame(dtm.todense()).sample(5000,random_state=123)
sample_dtm = sample.values  

# Set up "vocab", a list of feature names
vocab = np.array(vec.get_feature_names_out())

# Set up empty list to collect diagnostics
records = []
for top in range(40, 151, 10): 
    print(f"Fitting {top} topics")
    record = {'topics':top}
    # Fill in missing parameters
    lda = LDA(n_components=top,
              doc_topic_prior=25/top,
              topic_word_prior=0.15,
              n_jobs = -1,
              random_state=123)

    lda.fit(sample_dtm)# Add code to fit the the model
    
    ## Add code to compute umass, add mean umass to "record", and collect record in list
    umass = metric_coherence_gensim(measure="u_mass",
                                    top_n=5,
                                    topic_word_distrib=lda.components_,
                                    dtm=sample_dtm,
                                    vocab=vocab,
                                    texts=None)
    record['mean_umass'] = np.mean(umass)
    records.append(record)
```

```python
import matplotlib.pyplot as plt 
diagnostics = pd.DataFrame(records)
display(diagnostics)
fig, ax = plt.subplots(figsize=(8,5)) 

ax.plot(diagnostics['topics'], diagnostics['mean_umass'], marker='o', color='blue', linestyle='--', label='mean u_mass') 

ax.set_title("Topic Coherence (u_mass) vs Number of Topics") 
ax.set_xlabel("Number of Topics") 
ax.set_ylabel("Mean u_mass Coherence", color='black') 
ax.tick_params(axis='y', labelcolor='black') 

# best topic idx
best_idx = diagnostics['mean_umass'].idxmax() 

best_topics = diagnostics.loc[best_idx, 'topics'] 
best_score = diagnostics.loc[best_idx, 'mean_umass'] 

ax.scatter(best_topics, best_score, color='red', s=100, label=f'Best: {best_topics} topics') 

ax.legend() 
plt.show() 

print(f"Based on the plot, the number of topics that maximizes coherence is {best_topics} with mean u_mass = {best_score:.4f}")
```

## Topic Count Selection
Coherence scores were relatively stable across the tested range. Although 130 topics produced the strongest mean `u_mass` value, the notebook uses 50 topics as a more interpretable final specification while remaining close to the best coherence scores.

```python
# Fit the final LDA model
final_lda = LDA(n_components=50,
                doc_topic_prior=25/100,
                topic_word_prior=0.15,
                n_jobs=-1,
                random_state=123)
# Fit final model and transform the DTM
final_lda.fit(np.array(dtm.todense()))  # full data dtm
topics = final_lda.transform(np.array(dtm.todense()))
```

## Inspect Representative Topic Words

```python
import random
topn = 5

# You need these two objects:
vocab = np.array(vec.get_feature_names_out()) 
top_word = final_lda.components_

# Insert code to perform the intrusion task
def get_topic_words(topic, top_word, vocab, topn=5):
    """
    return top n words of the given topic
    """
    top_indices = top_word[topic, :].argsort()[-topn:][::-1]
    return vocab[top_indices]

n_topics = top_word.shape[0]

for top in range(n_topics):
    words = get_topic_words(top, top_word, vocab, topn=topn)
    random_word = random.sample(list([w for w in vocab if w not in words]), 1)[0]
    print(f"Topic {top} words: {' | '.join(words)}\t\tIntruder: {random_word}")
```

## Evaluate Topic Informativeness with Lasso

```python
from sklearn.model_selection import train_test_split

Xtopic_train, Xtopic_test, Ytopic_train, Ytopic_test = train_test_split(topics, df['BHAR0_2'], test_size=0.2,random_state=123)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import Lasso

np.random.seed(123)

param_grid = {
    'random_state':[123],
    'alpha':np.random.uniform(0.00001, 0.02, 100)
}

lasso = Lasso()

rus = RandomizedSearchCV(
    estimator=lasso,
    param_distributions=param_grid,
    n_iter=50,  # 50 iterations
    scoring='neg_root_mean_squared_error',
    cv=5,
    random_state = 123,
    n_jobs = -1
)
rus.fit(Xtopic_train, Ytopic_train)
```

```python
import matplotlib.pyplot as plt


alphas = rus.cv_results_['param_alpha']         
neg_rmse = rus.cv_results_['mean_test_score']   


plt.figure(figsize=(8,5))
plt.scatter(alphas, neg_rmse, color='blue')
plt.xlabel('Alpha')
plt.ylabel('Validation neg_root_mean_squared_error')
plt.title('Validation RMSE vs. Alpha for Lasso')
plt.grid(True)
plt.show()
```

```python
print("Best alpha:", rus.best_params_['alpha'])
```

```python
best_lasso = rus.best_estimator_
coefs = best_lasso.coef_

num_nonzero = (coefs != 0).sum()
print("Number of non-zero coefficients:", num_nonzero)
```

```python
coefs = best_lasso.coef_

most_pos = coefs.argmax()
most_neg = coefs.argmin()

print(f"The most positive coefficient corresponds to topic {most_pos} and has a value {coefs[most_pos]:.6f}.")
print(f"The most negative coefficient corresponds to topic {most_neg} and has a value {coefs[most_neg]:.6f}.")
```

```python
# Positive words:
words = get_topic_words(most_pos, top_word, vocab, topn=5)  # format as list
print(f"Topic {most_pos} (most positive) words: {'|'.join(words)}")

# Negative words:
words = get_topic_words(most_neg, top_word, vocab, topn=5)  # format as list
print(f"Topic {most_neg} (most negative) words: {'|'.join(words)}")
```

## Interpretation of Topic Coefficients
The strongest positive topic centers on EPS and adjusted operating performance, while the strongest negative topic centers on net loss language. This pattern is directionally intuitive because profit-oriented disclosure tends to align with favorable return reactions, while explicit loss language tends to align with weaker returns.

## Classification Models for Large Positive Returns

```python
YLR = (df['BHAR0_2']>5).astype(int)
train_Xdtm, test_Xdtm, train_Xtopic, test_Xtopic, train_YLR, test_YLR = train_test_split(np.asarray(dtm.todense()),topics,YLR,train_size=0.80,random_state=123)
```

```python
# Compare baseline classification models
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.naive_bayes import GaussianNB as NBC
from lightgbm import LGBMClassifier as LGBM
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

tree = DTC(max_depth = 25, max_features = 'sqrt', class_weight='balanced', random_state = 123)
tree.fit(train_Xtopic, train_YLR)
report_test = classification_report(test_YLR,tree.predict(test_Xtopic), output_dict=True, digits=3)
report_train = classification_report(train_YLR,tree.predict(train_Xtopic), output_dict=True, digits=3)

d = {
    'train_set': {'macro_f1': report_train["macro avg"]["f1-score"]},
    'test_set': {'macro_f1': report_test["macro avg"]["f1-score"]},
}
print('decision tree classification report')
print(pd.DataFrame.from_dict(d, orient='index'))
print('\n')

nbc = NBC()
nbc.fit(train_Xtopic, train_YLR)
report_test = classification_report(test_YLR,nbc.predict(test_Xtopic), output_dict=True, digits=3)
report_train = classification_report(train_YLR,nbc.predict(train_Xtopic), output_dict=True, digits=3)
d = {
    'train_set': {'macro_f1': report_train["macro avg"]["f1-score"]},
    'test_set': {'macro_f1': report_test["macro avg"]["f1-score"]},
}
print('naive bayes classification report')
print(pd.DataFrame.from_dict(d, orient='index'))
print('\n')

lgbm = LGBM(reg_alpha=0.02,random_state=123,n_jobs=-1,verbosity=-1)
lgbm.fit(train_Xtopic, train_YLR)
report_test = classification_report(test_YLR,lgbm.predict(test_Xtopic), output_dict=True, digits=3)
report_train = classification_report(train_YLR, lgbm.predict(train_Xtopic), output_dict=True, digits=3)

d = {
    'train_set': {'macro_f1': report_train["macro avg"]["f1-score"]},
    'test_set': {'macro_f1': report_test["macro avg"]["f1-score"]},
}
print('light GBM classification report')
print(pd.DataFrame.from_dict(d, orient='index'))

comments = '''
For the decision tree, the overfitting problem is obvious: training set has a much better performance than testing set in terms of macro f1. `max_depth` = 25 is too large and we can explore smaller choices later.
For the gaussian naive bayes model, the train and test sets have similar performance. This model is simple and not prone to overfit. 
LightGBM also overfits the training set. Although this model is an ensemble model, its performance is not very satisfying. Fine-turning is necessary.
'''
print(comments)
```
